```python
import os
from pathlib import Path
import pandas as pd
from textblob import TextBlob
import re
```

```python
current_dir = Path(os.getcwd()).absolute()
data_dir = current_dir.joinpath('data')

trump = pd.read_csv(data_dir.joinpath('hashtag_donaldtrump.csv'))
biden = pd.read_csv(data_dir.joinpath('hashtag_joebiden.csv'))
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) 

def get_tweet_sentiment(tweetPolarity): 
      
        # create TextBlob object of passed tweet text 

        if tweetPolarity > 0: 
            return 'positive'
        elif tweetPolarity == 0: 
            return 'neutral'
        else: 
            return 'negative'
        
def get_tweet_polarity(tweet): 
      
        # create TextBlob object of passed tweet text 
        analysis = TextBlob(tweet) 
        # set sentiment 
        return analysis.sentiment.polarity
```

```python
trump['cleantweet'] = trump.apply(lambda row: clean_tweet(str(row['tweet'])), axis = 1)
biden['cleantweet'] = biden.apply(lambda row: clean_tweet(str(row['tweet'])), axis = 1)
```

```python
trump['score'] = trump.apply(lambda row: get_tweet_polarity(str(row['tweet'])), axis = 1)
trump['sentiment'] = trump.apply(lambda row: get_tweet_sentiment(row['score']), axis = 1)

biden['score'] = biden.apply(lambda row: get_tweet_polarity(str(row['tweet'])), axis = 1)
biden['sentiment'] = biden.apply(lambda row: get_tweet_sentiment(row['score']), axis = 1)
```

```python
trumpclean = trump.filter(['created_at','cleantweet','score', 'sentiment'], axis=1)
bidenclean = biden.filter(['created_at','cleantweet','score', 'sentiment'], axis=1)

trumpclean.to_excel(data_dir.joinpath('cleantrump.xlsx'), index = False)
bidenclean.to_excel(data_dir.joinpath('cleanbiden.xlsx'), index = False)
```
