```python
"""
This function prints out a row in a more human readable format. It also converts Pounds to U$. 
"""

def printRow(row):
    row = rawdata.iloc[row] #This function pulls out the row. Iloc uses the row and column numbers to pull data from a
                            # pandas dataframe.
    counter = 0;
    badchars = str.maketrans("", "", "£,")

    for x in row:
        if list(rawdata.columns)[counter] == 'Price':
            print('Category: ',list(rawdata.columns)[counter],'\nValue: $', round(float(x.translate(badchars))*1.22, 2)) 
            # The original data was in pounds - this converts to US dollars.
        elif list(rawdata.columns)[counter] == 'Number Available In Stock': # Original data had value then condition. This prints it nicely.
            if isinstance(x, str):
                stock = x.split()
            else:
                stock = [x, '']
            print('Category: ',list(rawdata.columns)[counter],'\nValue: ', stock[0], 'Condition: ', stock[1].capitalize())
        else:
            print('Category: ',list(rawdata.columns)[counter],'\nValue: ', x)
        counter += 1
     
        
    return row
    
```

```python
"""
This sections iterates through and tests our Prices to make sure we had info for everything. About 15% of rows have
no price data we would need to take into account.
"""

NANcounter = 0
invalidCounter = 0
badchars = str.maketrans("", "", "£,")
for x in rawdata['Price']:
    if isinstance(x, str):
        if '-' not in x:
            print(round(float(x.translate(badchars))*1.22, 2))
        else:
            print('Invalid row: ', x)
            invalidCounter += 1
    else:
        NANcounter += 1
    
print('Number of NAN values: ',NANcounter)
print('Number of Invalid Rows: ', invalidCounter)
```

```python
from fuzzywuzzy import process
```

```python
"""
This function cleans our data. 
We discovered 7 NAN values in manufacturer and this replaces them with None.

"""
cleandata = rawdata
cleandata['Manufacturer'] = cleandata['Manufacturer'].replace(np.nan, 'None', regex=True)

choices = cleandata['Manufacturer'].unique()
topchoice = process.extract('Hornby', choices)
print(topchoice[0])
```

```python
newManufacturer = []
counter = 0
for x in cleandata['Manufacturer']:
    if isinstance(x, str):
        topChoice = process.extractOne(x, choices)
        #print(topChoice)
        newManufacturer.append(topChoice[0])
        counter += 1 #This function takes a while to run so I had it run 10 rows to start. It now serves as a clock.
        if (counter % 10) == 0:
            print(10000 - counter, 'Rows remaining.')
    else:
        print("Complicated!", x)
print(newManufacturer)
```
