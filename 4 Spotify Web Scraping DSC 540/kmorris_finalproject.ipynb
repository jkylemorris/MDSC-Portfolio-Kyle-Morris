```python
import pandas as pd
resultstring = json.dumps(result['tracks']['items'])
resultdf = pd.read_json(resultstring)
cleandf = resultdf
cleandf.rename(columns = {'name': 'song_name', 'duration_ms': 'duration_s'}, inplace = True) # This renames a few columns.
print(resultdf)

"""
This sections dumps our items string into a pandas DF, then creates the clean one where we will put all of our data.
"""
```

```python
resultdf['album'][0]['album_type'] # Album type (single, album, etc.)
resultdf['album'][0]['name'] # Album name.
resultdf['album'][0]['release_date'] # Release Date
resultdf['album'][0]['total_tracks'] # Total tracks

"""
More exploration. This is how I broke down the "album" section so that we could extract the data within.
"""
```

```python
albumtype = []
albumname = []
albumreleasedate = []
albumtotaltracks = []
for item in resultdf['album']:
    
    albumtype.append(item['album_type'])
    albumname.append(item['name'])
    albumreleasedate.append(item['release_date'])
    albumtotaltracks.append(item['total_tracks'])

"""print(albumtype)
print(albumname)
print(albumreleasedate)
print(albumtotaltracks)"""
cleandf['album_type'] = albumtype
cleandf['album'] = albumname
cleandf['album_release_date'] = albumreleasedate
cleandf['album_track_total'] = albumtotaltracks

"""
This is the big transformation. Pulling the 'album' data, creating a few lists and then importing that data back into our
clean dataframe. 

"""
```

```python
#artists = resultdf['artists'][0][0]['name'] 
def stripArtists(coolname):
    artists = ""
    x = 0
    while x < len(coolname):
        artists = artists + coolname[x]['name'] + " & "
        x += 1

    artists = artists[:-3]
    print(artists)
    return artists

"""
I wrote this function to strip the artists out. We don't know if a song has 1 or 17 artists, so this function tests for it!
It takes each individual artist listed and constructs a string. It results in Artist 1 & Artist 2 & etc. Once it's finished, it
strips the " & " from the end to tidy up. Coolname was a placeholder but it stuck.

"""
```

```python
cleanArtists = []
for item in resultdf['artists']:
    #print(item)
    """
    This portion sends each query result to give us the clean Artist field. It cleans up the Artist field instead of being
    nested.
    """
    cleanArtists.append(stripArtists(item))

cleandf['artists'] = cleanArtists
cleandf['duration_s'] = round((cleandf['duration_s'] / 1000), 2)
"""
This last bit converts the duration of the song in milliseconds into the duration of the song in seconds. A far more useful
stat, I think!

"""
```
